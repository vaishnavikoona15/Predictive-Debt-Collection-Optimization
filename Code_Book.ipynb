```python
# Create a box plot for the non-null columns

for column in null_columns_train:
    
    print(column)
    
    sns.boxplot(x=train_data[column].dropna(), color='red')

    plt.xlabel('Values')
    plt.title('Box Plot of Non Null Values')
    plt.show()
```

```python
# Calculate the mean, median and mode for each of the following columns.

for column in null_columns_train:
    print(f"{column} \n Mean: {train_data[column].dropna().mean()} \n Median: {train_data[column].dropna().median()} \n Mode: {train_data[column].dropna().mode()}")
    
```

```python
# As the distribution is skewed, using median imputation would be better for handling the null values in this column.
train_data['Max_Perc_Def_Chg_Pending'] = train_data['Max_Perc_Def_Chg_Pending'].fillna(train_data['Max_Perc_Def_Chg_Pending'].median())


# Missing Value Indicator for this column as huge amount of data is missing in this column.
train_data['Max_Ratio_OC_Pending_POS_Missing'] = train_data['Max_Ratio_OC_Pending_POS'].isnull().astype(int)


# Median Imputation as this column contains outliers.
train_data["N_PosBkt_L3m"] = train_data["N_PosBkt_L3m"].fillna(train_data["N_PosBkt_L3m"].median())


# Missing Value Indicator for this column as huge amount of data is missing in this column.
train_data["CoAp_Income_missing"] = train_data["CoAp_Income"].isnull().astype(int)


# Median Imputation as the data in this column is highly skewed.
train_data["Max_DPD_L3m"] = train_data["Max_DPD_L3m"].fillna(train_data["Max_DPD_L3m"].median())


# Median imputation as this column contains outliers.
train_data["Perc_Paymode_Cheq_Fail"] = train_data["Perc_Paymode_Cheq_Fail"].fillna(train_data["Perc_Paymode_Cheq_Fail"].median())


# Assumption: The null values represent that there has not been a customer visit done by a field representative and it has not been recorded
train_data["Total_Field_Trails"] = train_data["Total_Field_Trails"].fillna(0)


# Assumption: The null values represent 0, as the cases have been resolved but have not been recorded.
train_data["Total_Resolved"] = train_data["Total_Resolved"].fillna(0)


# Assumption: The column's missing values indicate no other loans so imputing them with 0.
train_data["Max_Loan_Balance_Others"] = train_data["Max_Loan_Balance_Others"].fillna(0)


# REPLACING THE NULL VALUES IN Max_Ratio_OC_Pending_POS AND CoAp_Income 

train_data["Max_Ratio_OC_Pending_POS"] = train_data["Max_Ratio_OC_Pending_POS"].fillna(train_data["Max_Ratio_OC_Pending_POS"].dropna().median())
train_data["CoAp_Income"] = train_data["CoAp_Income"].fillna(train_data["CoAp_Income"].dropna().median())
```

# Model Testing

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, accuracy_score

# Defining the features and the target variable 
X = train_data.drop('Target_Flag', axis=1)  # Features
y = train_data['Target_Flag']  # Target variable

# Spliting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Defining hyperparameters for the LightGBM model
params = {
    'objective': 'binary',
    'boosting_type': 'gbdt',
    'metric': 'auc',  
    'num_leaves': 31,  
    'learning_rate': 0.05,  
    'feature_fraction': 0.9,
}

# Creating a LightGBM dataset for training
train_data = lgb.Dataset(X_train, label=y_train)

# Training the LightGBM model
num_round = 100  # Adjust as needed
bst = lgb.train(params, train_data, num_round)

# Making predictions on the test data
y_pred = bst.predict(X_test)

# Converting probabilities to binary predictions (0 or 1)
y_pred_binary = [1 if p >= 0.5 else 0 for p in y_pred]

# Calculating accuracy and AUC-ROC score
accuracy = accuracy_score(y_test, y_pred_binary)
roc_auc = roc_auc_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print(f"AUC-ROC Score: {roc_auc:.2f}")
```

"LightGBM" model with "Train_Test_Split" was found to give highest AUC ROC Score of 0.70 among the experimented models, when compared by splitting the training dataset.

# Final Model: LightGBM

```python
# Defining the X_train, X_test and y_train

X_train = X
y_train = y
X_test = test_data

# Defining hyperparameters for the LightGBM model
params = {
    'objective': 'binary',
    'boosting_type': 'gbdt',
    'metric': 'auc',  
    'num_leaves': 31,  
    'learning_rate': 0.05,  
    'feature_fraction': 0.9,
}

# Creating a LightGBM dataset for training
train_data = lgb.Dataset(X_train, label=y_train)

# Training the LightGBM model
num_round = 100  # Adjust as needed
bst = lgb.train(params, train_data, num_round)

# Making predictions on the test data
y_pred_proba = bst.predict(X_test)

# Converting probabilities to binary predictions (0 or 1)
y_pred = [1 if p >= 0.5 else 0 for p in y_pred_proba]
```
